# Mental Health Digital Shadows — Data Pooling System Dashboard

An interactive browser for the **MHDS dataset**: 75,000 LLM-generated responses across **15 large language models**, each persona conditioned on sociodemographic and psychological attributes (Big Five OCEAN, DASS-21 severity for depression, anxiety, and stress). 

📊 **Dataset:** [github.com/MassimoStel/MHDS](https://github.com/MassimoStel/MHDS)  
🤖 **Lab:** [CogNosco Lab](https://cognosco.dipsco.unitn.it/), University of Trento  
📄 **Paper:** *Digital shadows in mental health map how LLMs simulate depression, anxiety, and stress through language and psychometrics* — Franchino, Rizzi, De Duro, Ardebili, Stella (in submission) | [Preprint (PsyArXiv)](https://osf.io/7zhvr_v1)

### How to use

1. **Runtime → Run all** (~3 min on first launch; the data are downloaded to your Drive).
2. Select a Model Name and filter personas in the left panel, click **Search Personas**, then explore each persona's six topic responses, ERT word list, and DASS-21 scores in the right panel.
3. Use **Download** to export the filtered pool as XLSX or PKL.
